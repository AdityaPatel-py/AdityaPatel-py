```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['normalized-losses'])
plt.show()
```

```python
sns.distplot(auto['engine-size'])
plt.show()
```

```python
sns.distplot(auto['city-mpg'])
plt.show()
```

```python
sns.distplot(auto['city-mpg'], kde = False, rug = True)
plt.show()
```

```python
sns.jointplot(x = auto['engine-size'], y = auto['horsepower'])
plt.show()
```

```python
sns.pairplot(auto[['normalized-losses','engine-size','horsepower']])
```

```python
sns.stripplot(x = auto['fuel-type'], y = auto['horsepower'])
```

```python
sns.swarmplot(x = auto['fuel-type'], y = auto['horsepower'])
```

```python
sns.boxplot(x= auto['num-of-doors'], y = auto['horsepower'], hue = auto['fuel-type'])
```

```python
sns.barplot(x= auto['body-style'], y = auto['horsepower'], hue = auto['engine-location'])
```

```python
sns.countplot(x = auto['body-style'], hue = auto['engine-location'])
```

```python
sns.pointplot(x = auto['fuel-system'], y = auto['horsepower'], hue = auto['num-of-doors'])
```

```python
sns.catplot(x = 'fuel-type', y = 'horsepower', hue = 'num-of-doors', col = 'engine-location', data = auto, kind = 'strip')
```
